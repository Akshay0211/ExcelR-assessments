# Anova ftest statistics: stats.f_oneway(column-1,column-2,column-3,column-4)

```python
p_value=stats.f_oneway(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2],df.iloc[:,3])
p_value
```

P value of the One way Anova test is < α
H1 is accepted.
Mean TAT for 4 laboratories not equal (There is a significance difference between TAT of the laboratories)
