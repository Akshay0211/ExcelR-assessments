```python
books=pd.read_csv('book (1).csv',encoding='Latin1')
books
```

```python
books1=books.iloc[:,1:]
books1
```

```python
# Sort by User IDs
books1.sort_values(['User.ID'])
```

```python
# number of unique users in the dataset
len(books1['User.ID'].unique())
```

```python
# number of unique books in the dataset
len(books1['Book.Title'].unique())
```

```python
# converting long data into wide data using pivot table
books2=books1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books2
```

```python
# Replacing the index values by unique user Ids
books2.index=books1['User.ID'].unique()
books2
```

```python
# Impute those NaNs with 0 values
books2.fillna(0,inplace=True)
books2
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(books2.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim1=pd.DataFrame(user_sim)
user_sim1
```

```python
user_sim1.index=books1['User.ID'].unique()
user_sim1.columns=books1['User.ID'].unique()
user_sim1
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim1
```

```python
# Most Similar Users
user_sim1.idxmax(axis=1)
```
